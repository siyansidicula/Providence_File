```python
#Training-----------

#1.Load the DATA(excel sheet)
import pandas as pd
file=pd.read_csv("/content/13 MLP winequality-red.csv")
x=file[["total sulfur dioxide","density","pH","sulphates","alcohol"]]
y=file["quality"]

#Cross Validation
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2)


#2.Load the MODEL (percepton)
from sklearn.linear_model import Perceptron
model=Perceptron()

#3.fit DATA into MODEL
model.fit(xtrain,ytrain)

#Validation Testing
op=model.predict(xtest)
from sklearn.metrics import accuracy_score
score=accuracy_score(ytest,op)
print("Accuracy=",score*100)

#Testing-----------predict
result=model.predict([[1.5,2.5,3.5,4.5,5.5]])
print(result)
```

```python
#Training-----------

#1.Load the DATA(excel sheet)
import pandas as pd
file=pd.read_csv("/irisexcel.csv")
x=file[["sepal_length","sepal_width","petal_length","petal_width"]]
y=file["species"]

#Cross Validation
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2)


#2.Load the MODEL (percepton)
from sklearn.linear_model import Perceptron
model=Perceptron()

#3.fit DATA into MODEL
model.fit(xtrain,ytrain)

#Validation Testing
op=model.predict(xtest)
from sklearn.metrics import accuracy_score
score=accuracy_score(ytest,op)
print("Accuracy=",score*100)

#Testing-----------predict
result=model.predict([[1.5,2.5,3.5,4.5]])
print(result)
```
